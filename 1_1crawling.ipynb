```python
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Trident/7.0; rv:11.0) like Gecko'}
response= requests.get('https://www.melon.com/chart/index.htm', headers=header)
html=response.text
soup=BeautifulSoup(html, 'html.parser')

title=soup.find_all("div", {"class":"ellipsis rank01"})
singer=soup.find_all("div",{"class":"ellipsis rank02"})
real_title=[i.find('a').text for i in title]
real_singer=[i.find('a').text for i in singer]

for r in range(100):
    print(' %2d위 : %s -  %s'%(r+1, real_singer[r], real_title[r]))
```

```python
Top100List=[]
for i in range(100):
    temp=[]
    temp.append(real_singer[i])
    temp.append(real_title[i])
    Top100List.append(temp)
Top100=pd.DataFrame(Top100List)
Top100.columns=['singer','title']
Top100.head()
Top100.to_csv('MelonTop100.csv', encoding='cp949')
```

```python
title= 'MelonTop100-2'
f=open(f'{title}.csv','w',encoding='cp949',newline='')
csvWriter=csv.writer(f)
for i in Top100List:
    csvWriter.writerow(i)
f.close()
```
